```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = pd.read_csv('kmean.csv')
```

```python
year = np.zeros((len(data)))
for i in range(len(data)):
    if 0<=i<=101:
        year[i]=int(2010)
    elif 101<i<= 247:
        year[i]=int(2011)
    elif 247<i<=438:
        year[i]=int(2012)
    elif 438<i<=596:
        year[i]=int(2013)
    elif 596<i<=760:
        year[i]=int(2014)
    elif 760<i<=894:
        year[i]=int(2015)
    elif 894<i<=1027:
        year[i]=int(2016)
    elif 1027<i<=1120:
        year[i]=int(2017)
    else:
        year[i]=int(2018)
        
```

```python
data['YEAR']= year
```
