```python
import requests

from bs4 import BeautifulSoup

import newspaper
from newspaper import Article

import pandas as pd
```

```python
hiv_scrape = pd.DataFrame(columns=['DATE','AUTHOR','IMAGE','TEXT','KEYWORDS', 'SUMMARY'])
```

```python
hiv_scrape
```

```python
url_archive = 'https://timesofindia.indiatimes.com/'
```

```python
#https://timesofindia.indiatimes.com/2010/4/30/archivelist/year-2010,month-4,starttime-40298.cms

index = 40302 #40179
count = 21 #0

for year in range(2010, 2019): #2010,2019
    
    for month in range(1,13): #1,13
        
        if month == 1 | month == 3 | month == 5 | month == 7 | month == 8 | month == 10 | month == 12 :
            date_max = 31
        elif month == 2 & year % 4 == 0:
            date_max = 29
        elif month == 2 & year % 4 != 0:
            date_max = 28
        else:
            date_max = 30
            
        for date in range(1, date_max+1): 
        
            url = url_archive + str(year) + '/' + str(month) + '/' + str(date) + '/archivelist/year-' + str(year) + ',month-' + str(month) + ',starttime-' + str(index) + '.cms'
        
            page = requests.get(url)
            soup = BeautifulSoup(page.text, 'html.parser')
            
            tags = soup.find_all('a')
            
            # Extracting URLs from the attribute href in the <a> tags.
            for tag in tags:
                try:
                    if (('hiv' in tag.get('href')) | ('Hiv' in tag.get('href')) | ('HIV' in tag.get('href'))) & ('archive' not in tag.get('href')):
                        
                        #print(tag.get('href'))
                        date_article = str(date) + '-' + str(month) + '-' + str(year)
                        #print(date_article)                      
                        article = Article(tag.get('href'))
                        article.download()
                        article.html
                        article.parse()
                        
                        article.text
                        article.nlp()
                        key = article.keywords
                        
                        print(date_article)
                        print(article.keywords)
                        
                        if (('hiv' in key) | ('Hiv' in key) | ('HIV' in key)):
                        
                            #date
                            hiv_scrape.at[count, 'DATE'] = date_article
                        
                            #author
                            hiv_scrape.at[count, 'AUTHOR'] = article.authors
                        
                            #image
                            hiv_scrape.at[count, 'IMAGE'] = article.top_image
                        
                            #text
                            hiv_scrape.at[count, 'TEXT'] = article.text

                            #summary / keywords                        
                            hiv_scrape.at[count, 'KEYWORDS'] = article.keywords
                            hiv_scrape.at[count, 'SUMMARY'] = article.summary
                        
                            print(hiv_scrape.loc[[count]])
                            count += 1
                        
                except:
                    pass 
                
            print ('count : ' + str(count) + '\n' + 'index : ' + str(index) + '\n')
            
            index += 1
```

```python
hiv_scrape.to_csv('hiv_scrape.csv')
```
