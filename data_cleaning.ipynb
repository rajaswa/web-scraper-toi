```python
import pandas as pd
```

```python
toi_2010_13 = pd.read_csv('hiv_scrape_2010-13.csv')
toi_2014_15 = pd.read_csv('hiv_scrape_2014-15.csv')
toi_2016 = pd.read_csv('hiv_scrape_2016.csv')
toi_2017 = pd.read_csv('hiv_scrape_2017.csv')
toi_2018 = pd.read_csv('hiv_scrape_2018.csv')
```

```python
def concat(source, target):
    return (pd.concat([source, target], ignore_index=True))
```

```python
toi_2010_18 = concat(toi_2010_13, (concat(toi_2014_15, (concat(toi_2016, (concat(toi_2017, toi_2018)))))))
```

```python
len(toi_2010_18)
```

```python
toi_2010_18 = toi_2010_18.drop('Unnamed: 0', 1)
```

```python
toi_2010_18.to_csv('toi_2010_18.csv', index = False)
```
