```python
import requests
from bs4 import BeautifulSoup
import newspaper
from newspaper import Article
```

# Example with one article

```python
url = 'https://timesofindia.indiatimes.com/topic/Hiv/news'
```

```python
url_head = 'https://timesofindia.indiatimes.com'
```

```python
page = requests.get(url)
```

```python
print(page.text)
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
tags = soup.find_all('a')
```

```python
article_url = []

# Extracting URLs from the attribute href in the <a> tags.
for tag in tags:
    try:
        if ('hiv' in tag.get('href')):
            article_url.append(url_head + str(tag.get('href')))
            print(tag.get('href'))
    except:
        pass
```

```python
article = Article(article_url[0])
```
